## 1. 편향(Bisa)

- 모델이 너무 단순해서 데이터의 진짜 관계를 찾아내지 못한 현상
- 과소적합
- 예
    - 키로 몸무게를 예측
    - 실제로는 키가 커질수록 체중이 증가하다가 어느 정도 수렴하는 곡선 관계
- 이 문제를 선형회귀(직선)모델을 쓰면 이 곡선을 패턴을 찾아내지 못하고,
- 오차(실제값과 추정선 사이 거리)가 커지는 문제가 생깁니다.
- 학습 데이터에도, 새 데이터에도 모두 성능이 좋지 않은 상태
    - 이 상태를 언더비팅, 편향이 높다고 표현합니다.

## 2. 분산(Variance)

- 모델이 너무 복잡해서 학습 데이터는 너무 잘 맞추지만, 새 데이터에는 성능이 안좋게 나오는 현상
    - 오버피팅이라고 합니다.
- 분산이 높다라고 표현합니다.

## 3. 편향과 분산의 트레이드오프

- 일반적으로 편향과 분산은 하나가 줄어들수록 다른 하나는 늘어나는 관계
- 좋은 모델이라고 하는 것은 적절한 복잡도를 찾아서 편향과 분산을 낮춰야 하는 것이 목적입니다.
- 너무 단순한 모델은 높은 편향
- 너무 복잡한 모델은 높은 분산
- 그래서 적당한 모델을 찾는 것이 목표입니다.
