# 피처 스케일링과 정규화

- 피처 스케일링의 방법으로는 표준화와 정규화가 있습니다
- 표준화는 피처 각각이 평균이 0이고, 분산이 1인 가우시안 정규 분포를 가진 값으로 변환하는 것을 의미합니다
- 원래 값에서 피처 x의 평균을 뺀 값을 피처 x의 표준편차로 나눈 값을 계산할 수 있습니다

---

- 일반적으로 정규하는 서로 다른 피처의 크기를 통일하기 위해 크기를 변환해주는 개념입니다
- 예를 들어 피처 A는 거리를 나타내는 변수로서 값이 0 ~ 100KM로 주어지고
- B는 금액을 나타내는 속성으로 값이 0 ~ 100,000,000,000원으로 주어진다면
- 이 변수를 모두 동일한 크기 단위로 비교하기 위해 값을 모두 최소 0 ~ 최대 1의 값으로 변환하는 것을 말합니다.
- 즉 개별 데이터의 크기를 모두 똑같은 단위로 변경하는 것 입니다.
-  원래 값에서 피처 x의 최솟값을 뺀 값을 피처 x의 최댓값과 최솟값의 차이로 나눈 값으로 변환할 수 있습니다

- 혼선을 방지하기 위해 일반적인 의미의 표준화와 정구화를 피처 스케일링으로 통칭하고
- 선형대수 개념의 정규화를 벡터 정규화로 지칭하겠습니다
- 먼저 피처 스케일링 클래스인 StandardScaler와 MinMaxXScaler를 알아보겠습니다

# StandardScaler

- 표준화를 쉽게 지원하기 위한 클래스입니다
- 개별 피처를 평균이 0이고, 분산이 1인 값으로 변환해줍니다.
- 서포트 벡터 머신이나 선형 회귀, 로지스틱 회귀는 데이터가 가우시안 분포를 가지고 있다고 가정하고 구현됐기 때문에 사전에 표준화를 적용하는 것은
- 예측 향상에 중요한 요소가 될 수 있습니다
- 어떻게 데이터 값을 변환하는지 데이터 세트로 확인해 보겠습니다

````python
# 데이터 불러오기
from sklearn.datasets import load_iris
import pandas as pd

# 붓꽃 데이터를 프레임 형태로 변환
iris = load_iris()
iris_data = iris.data
df = pd.DataFrame(iris_data, columns=iris.feature_names)
df
````

````python
print(f'feature들의 평균 값')
print(df.mean())
print(f'\feature들의 분간 값')
print(df.var())
````

- 이제 standardscaler를 사용해서 각 피처를 한번에 표준화를 진행해 보겠습니다
- standardscaler를 생성한 후에 fit() sransform() 메서드에 변환 대상 데이터를 입력하고 호출하면 간단하게 변환됩니다

````python
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체 생성
scaler = StandardScaler()

# StandardScaler로 데이터 세트 변환, fit() transform() 호출
scaler.fit(df)
iris_scaled = scaler.transform(df)

# transform() 시 스케일 변환된 데이터 세트가 numpy로 반환돼 이를 데이터 프레임으로 변환
sclaer_df = pd.DataFrame(data=iris_scaled, columns=iris.feature_names)
sclaer_df
````

````python
print(f'feature들의 평균 값')
print(sclaer_df.mean())
print(f'\feature들의 분간 값')
print(sclaer_df.var())
````

- 모든 칼럼 값의 평균이 0 에 아주 가까운 값으로, 그리고 분산은 1에 아주 가까운 값으로 변환됐음을 알 수 있습니다

# MinMaxScaler

- 데이터값을 0과 1 사이의 범위 값으로 변환(음수 값이 있으면 -1 에서 1값으로 변환합니다)
- 데이터의 분포가 가우시안 분포가 아닐 경우에 min, max scaler를 적용해 볼 수 있습니다

````python
from sklearn.preprocessing import MinMaxScaler

# 객체 생성
minmax = MinMaxScaler()

# fit(), transform() 호출
minmax.fit(df)
minmax_scaler = minmax.transform(df)

# 데이터 프레임으로 저장
minmax_df = pd.DataFrame(data=minmax_scaler, columns=iris.feature_names)
minmax_df
````

````python
print(f'feature들의 최솟값')
print(minmax_df.min())
print(f'\feature들의 최댓값')
print(minmax_df.max())
````

# 학습데이터와 테스트 데이터의 스케일링 변환 시 유의점

위와 같은 스케일링 객체를 이용해 데이터의 스케일링 변환 시
- fit(), tranform(), fit_transform() 메서드를 이용합니다

```
일반적으로 fit()은 데이터 변환을 위한 기준 정보 설정(예를 들어 데이터 세트의 최댓값/최솟값 설정 등)을 하며,
transform()은 이렇게 설정된 정보를 이용해 데이터를 변환합니다.
그리고 fit_transform()은 위 두 과정을 한번에 적용하는 기능을 수행합니다

스케일러 객체를 이용해 학습 데이터 세트로 fit(), transform()를 적용하면 테스트 데이터 세트로는 다시 fit()을 수행하지 않고
학습 데이터 세트로 fit()을 수행한 결과를 이용해 transform() 변환을 적용해야 한다는 것입니다

즉 학습 데이터로 fit()이 적용된 스케일링 기준 정보를 그대로 테스트 데이터에 적용해야 하며
그렇지 않고 테스트 데이터로 다시 새로운 스케일링 기준 정보를 만들게 되면 학습 데이터와 테스트 데이터의 스케일링 기준 정보가 서로 달리지기 때문에
올바른 예측 결과를 도출하지 못할 수 있습니다

예제를 통해 알아보겠습니다
```

````python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# 학습 데이터는 0부터 10까지
# 테스트 데이터는 0부터 5까지 값을 가지는 데이터 생성
# scaler 클래스의 fit() 함수와 tranform() 은 2차원 이상 데이터만 가능하므로 reshape(-1,1)로 차원 변경

train_array = np.arange(0, 11).reshape(-1, 1)
test_array = np.arange(0, 6).reshape(-1, 1)
````

- 학습 데이터에는 0부터 10까지 값을 가지는데
- 이 객체에 fit() 적용하면 최솟값0, 최댓값10이 설정되며 1/10 스케일링이 적용됩니다
- 이제 transform() 을 호출하면 1/10 스케일링으로 학습 데이터를 변환하게 되며 원본 데이터1은 0.1로 2는 0.2로 10은 1로 변환됩니다

````python
# 객체 저장
scaler = MinMaxScaler()

# fit() 활용 : 객체가 저장된 변수에 기준 정보를 학습 데이터로 파라미터 저장
# 데이터의 최솟값이 0, 최댓값이 10으로 설정
scaler.fit(train_array)

# 1/10 scale로 데이터 변환. 원본 10-> 1로 변환
train_scaled = scaler.transform(train_array)
````

````python
train_scaled
````

````python
print(f'원본 train_array 데이터 : {np.round(train_array.reshape(-1), 2)})')
print(f'train_array 데이터 : {np.round(train_scaled.reshape(-1),2)}')
````

- 이번에는 테스트 데이터 세트를 변환하는데, fit()을 호출해 스케일링 기준 정보를 다시 적용한 뒤 transform()을 수행한 결과를 확인해봅시다

````python
# MinMaxScaler에 test_array를 fit()하게 되면 원본 데이터의 최솟값이0, 최댓값이 5
scaler.fit(test_array)

# 1/5 scaler로 test_array 데이터 변환함. 원본 5 -> 1로 변환
test_scaled = scaler.transform(test_array)

# test_array의 scale 변환 출력
print(f'원본 test_array 데이터 : {np.round(test_array.reshape(-1), 2)})')
print(f'test_array 데이터 : {np.round(test_scaled.reshape(-1),2)}')
````

- 출력 결과를 확인하면 학습 데이터와 테스트 데이터의 스케일링이 맞지 않음을 알 수 있습니다.
- 앞서 학습 데이터로 동일한 값으로 변환되는 결과를 초래합니다
- 머신러닝 모델은 학습 데이터를 기반으로 학습되기 때문에 반드싵 ㅔ스트 데이터는 학습 데이터의 스케일링 기준에 따라야 하며, 테스트 데이터의 1값은 학습 데이터와 동일하게
- 0.1값으로 변환되어야 합니다
- 따라서 테스트 데이터에 다시 fit()을 적용해서는 안되며 학습 데이터로 이미 fit()이 적용된 scaler 객체를 이용해 transform()으로 변환해야 합니다

````python
scaler = MinMaxScaler()
scaler.fit(train_array)
train_scaled = scaler.transform(train_array)
print(f'원본 train_array 데이터 : {np.round(train_array.reshape(-1), 2)})')
print(f'train_array 데이터 : {np.round(train_scaled.reshape(-1),2)}')

# test_array에 scale은 변환을 할 때는 반드시 fit()을 호출하지 않고 transform()만으로 변환해야 함
test_scaled = scaler.transform(test_array)
print(f'\n원본 test_array 데이터 : {np.round(test_array.reshape(-1), 2)})')
print(f'test_array 데이터 : {np.round(test_scaled.reshape(-1),2)}')
````

```
fit_transform() 적용할 때도 만찬가지 입니다.
fit_transform()은 fit() transform()을 순차적으로 수행하는 메서드이므로 학습 데이터에서는 상관 없지만 테스트 데이터에서는 절대 사용해서는 안됩니다
이렇게 위 객체를 적용할 때 주의 사항이 발생하므로 학습과 테스트 데이터 세트로 분리하기 전에 먼저 전체 데이터 세트에 스케일링을 적용하 뒤
학습과 테스트 데이터 세트로 분하는 것이 더 바람직합니다

1. 가능하다면 전체 데이터의 스케일링 변환을 적용한 뒤 학습과 테스트 데이터로 분리
2. 1이 여의치 않다면 테스트 데이터 변환 시에는 fit()이나 fit_transform()을 적용하지 않고 학습 데이터로 이미
fit() 된 scale 객체를 이용해 transform()으로 변환합니다
```
