# 회귀분석의 사용목적

### (단순)선형회귀모델의 기본 공식

- $y = {\beta_0 + \beta_1x}$
    - y는 예측하려는 대상(종속, 타켓, 레이블)
    - x는 설명에 쓰일 하나의 숫자(독립, 특성, 피처)
    - ${\beta_0}$는 직선이 y축과 만나는 절편(intercept)
    - ${\beta_1x}$는 x가 1 증가할때 y가 얼만큼 변하는지 나타내는 기울기(coefficient)

image.png

### 공식의 의미

- 공부시간과 시험점수
- 점수 = 30 + 5 * 공부시간
- 절편 30점은 하루도 공부하지 않았을 때 평균 점수
- 기울기 5점 1시간 더 공부하면 평균적 5점이 오른다는 의미
- 시각적으로 산점도 위에 직성을 "가장 잘" 맞춰 그리는 것이 목표
- 가장 잘의 기준은 오차(실제 - 예측)의 제곱합(MSE)를 최소화하는 것

### 직선을 맞춘다는 것

- 직선은 두 숫자로 정해짐
    - 절편: y축에서 시작 위치
    - 기울기: x를 1만큼 옮길 때 y가 얼마나 변하는가?
- 그래프 위에 자를 올린 뒤
- 자를 위 아래, 기울기를 살짝 움직인다면
- 가장 많은 점에 가까워지는 위치를 찾는 과정이 "학습"

### 직선을 맞추는 과정 학습이라고 표현하는 이유

- 우리는 모델에게 절편과 기울기 값을 가르쳐주지 않음
- 모델은 데이터를 보고 스스로 이곳에 위치하는 선이 제일 적절하다고 판단하는 것
- 통계 용어로는 추정(estamate), 머신러닝 용어로는 학습(fit)이라고 함

### 어떻게 최적의 선을 찾을 수 있을까?

- 임의 위치에 직선을 일단 그려봄
- 절편=0, 기울기=1 이라고 가정
- 점마다 (실제값 - 직선이 예측한 값)의 차이를 구하는데, 이것을 오차라고 합니다.
- 오차를 한 번에 합쳐서 하나의 숫자로 만듬
    - 이것을 전체 오차의 평균으로 설정
- 이 떄 가장 일반적인 방법이 MSE(평균 제곱 오차)
- MSE를 머신러닝에서는 손실함수(Loss Function)
- 이 오차의 숫자가 최소가 될 때까지 절편과 기울기를 계속 움직이면서 값을 찾아갑니다.

### 회귀모델을 학습한다는 것

- 계수 ${\beta}$를 학습하는 것
- 계수 학습은 점과 직선 거리를 최소화하도록 절편과 기울기 위치를 자동으로 조절하는 과정
- 이 과정을 거치면 직선이 데이터의 '경향(trend)'를 기억하고
- 새로운 x값에도 y를 예측할 수 있게 됩니다.

## 손실함수와 설명력 지표:MSE, RMSE, R2

### RSS(Residual Sum of Squares)

- 각 관측값과 모델 예측값 간 차이(잔차)를 제곱해서 모두 더한 합
- 잔차: Residual
- RSS : Residual Sum of Squares
$$
? = \sum_{i=1}^{N} (y_i - \hat y_i)^2
$$

- N: 데이터 샘플 수
- y: 실제값(실제 집값, 실제 점수 등)
- y hat: 모델이 예측한 값
- 제곱하는 이유는?
    - 오차의 부호(양수/음수)를 없애기 위해

### MSE(Mean Sqaured Error, 평균제곱오차)

- 모델이 예측한 값과 살제 값 간 차이
- 평균 제곱 오차

$$
MSE = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat y_i)^2
$$

- 해석은?
    - 값이 작을수록 모델 예측이 실제에 가깝다는 뜻
- 문제가 있다면?
    - 음수를 해결할려고 하다보니 원래값보다 높은 값들이 나올 수 있습니다. 제곱을 했기 때문에
    - 이것을 해결할 떄 사용하는 것은?

### RMSE(Root Mean Sqaured Error, 평균 제곱근 오차)

- MSE에 제곱근을 취해, 원래 값과 같은 단위로 오차를 돌려받습니다
$$
RMSE = \sqrt{MSE}
$$

- 해석방법
    - 예를 들어 주택 가격 에측
    - RMSE가 5,000,000원이라면?
    - 내가 학습한 모델이 실제값과 평균적으로 500만원 정도 차이가 난다

### ${R}^2$ (결정계수)

- 모델이 전체 데이터 변동량 중 얼마나 설명하고 있는지를 비율로 나타낸 지표
- 0 ~ 1 사이 값을 가집니다.
- R2 값이 0인 경우, 모델이 실제 평균만큼도 설명을 못함
- R2 값이 1인 경우, 모델이 모든 변동을 완벽히 설명

$$
  R^2 = 1 - \frac{\sum_{i}(y_i - \hat y_i)^2}{\sum_{i}(y_i - \bar y)^2}
  $$

- 분모
- ${\bar y}$: 실제 값의 평균

### 오차와 잔차

### 오차(Error)

- 실제 관측값과 진짜 관계식의 차이
- Error = y - f(x)
- f(x)는 현실 속에 존재하는 완벽한 인과관계를 의미
- 하지만, 이론상으로만 존재할 뿐 현실적으로는 알애날 수 없음
- 데이터에 개입된 모든 불확실성(측정 오차, 숨겨진 변수, 노이즈 등)이 포함되어 있기 때문
- 모델로는 결고 완전히 줄일 수 없는 환계치가 존재

### 잔차(Residual)

- 실제 관측값과 모델 예측값의 차이
- Residual = y - yhat
- 학습한 모델은 우리 손으로 계산 가능한 관계식
- 모델 성능을 진단 및 개선이 가능
    - MSE, RSS, RMSE
    - 잔차를 줄이는 방향으로 치러를 추가학나, 모델 변경, 정규화 등을 진행
- 잔차로 모델을 진단, 개선하고 이론적으로 오차 개념으로 '이상적 한계'를 이해합니다.

## 비용함수(Cost Function)

$$
RSS = \sum_{i=1}^{N} (y_i - \hat y_i)^2
$$

$$
{\hat y} = {w_0 + w_1x}
$$

$$
\mathrm{RSS}(w_{0}, w_{1})
= \sum_{i=1}^{N} \bigl(y_{i} - (w_{0} + w_{1} x_{i})\bigr)^{2}
$$


- 비용함수는 모델의 예측괎과 실제값 간의 오차를 하나의 숫자로 요약해주는 함수
- 머신러닝에서는 이 비용함수를 최소화하는 것이 학습의 목표가 됩니다.(선형회귀함수에서는 MSE)

### 비용함수를 어떻게 최소화하는가?

- 최소자승법
    - 선형대수
- 경사하강법
    - 미분

## 다중선형회귀

- 현실적인 문제는 보통 여러 요인이 함께 작용하게 됩니다.
- 독립변수가 2개 이상 사용되면,
- 즉, 여러 개의 피처가 하나의 연속형 타겟을 예측할 때 사용하는 회귀모델

### 선형회귀모델의 기본가정

- 선형성 : x축이 늘어날 때마다 y축도 늘어나는 관계를 가져야함
- 독립성 : 각 관측치의 오차(잔차)가 서로 영향을 주지 않고, 독립적으로 발생해야 한다는 가정
    - 잔차 = 실제값 - 모델예측값
    - 잔차가 무작위로 흩어져 있으면 독립성 성립
    - 잔차가 비슷한 패턴을 보이면 독립성 깨짐
- 등분산성
- 정규성 : 예측하고 싶은 데이터가 정규분포를 이뤄야 함
- 다중공선성
    - 여러 독립변수들(피처)끼리 서로 강하게 (선형)상관되어 있는 현황
    - 예: 집값을 예측할 때 방 개수와 방 면적이 거의 비례한다면 이 둘은 서로 중복된 정보를 제공
    - 변수들 간 정보가 겹치게 되면 생기는 문제
        - 회귀계수의 분산이 커져 추정값이 불안정해지고,
        - 어떤 변수가 실제로 중요한지 판단하기 어려워집니다.
- 위 다섯 가지 가정이 충족될 때 선형회귀모델의 결과(계수, p값, 신뢰구간..)를 신뢰할 수 있습니다.

```python
from sklearn.datasets import fetch_california_housing  # 데이터 가져옥;
from sklearn.model_selection import train_test_split  # 분리
from sklearn.linear_model import LinearRegression  # 학습
from sklearn.metrics import mean_squared_error  # 평가
```

- MedInc: 중위소득
- HouseAge
- AveBedrms : 가구당 평균 침실 수
- AveOccup : 가구당 평균 거주자 수

```python
y.head()
```

```python
# 데이터 분리하기
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=96)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import r2_score

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'MSE={mse}')
print(f'RMSE={rmse}')
print(f'R2={r2}')  # 설명력
```

## 다중회귀분석에서 계수 해석 방법
- 계수의 해석은 "다른 모든 변수를 고정했을 때 특정 변수 X가 한 단위 변할 때 종속변수 y가 평균적으로 얼마나 변하는가?"
- 절편
    - 모든 x값이 0일 때의 예측값
- 기울기
    - x를 1만큼 증가시키면, y가 계수만큼 평균적으로 변화한다
    - 이 때 다른 독립변수 값을 모두 고정합니다.
- 예를 들어
    - 집값 = 1000 + (50 * 방갯수) + (20 * 평균)
    - 방 갯수가 1개 늘면, 다른 변수 고정 시 집값이 50만원 오름
    - 평수가 1평 늘면, 다른 변수 고정 시 집값이 20만원 오름
- 단위(Unit)에 주의해야 합니다
    - 각각 계수값은 독립변수의 단위에 종속됩니다.
    - 방개수 1개
    - 평수 1평
    - 소득 $10,000 단위 등

# 코드 익히기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing  # 데이터 가져오기
from sklearn.model_selection import train_test_split  # 분리
from sklearn.metrics import mean_squared_error, root_mean_squared_error, r2_score
```

```python
df = fetch_california_housing(as_frame=True)
X, y = df.data, df.target
```

```python
X.head()
```

```python
# 학습/테스트 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=95)
```

```python
# 모델 학습
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# RMSE, R2 평가
y_pred = model.predict(X_test)  # test값으로 예측시키기

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'MSE={mse}')
print(f'RMSE={rmse}')
print(f'R2={r2}')
```

```python

mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE={mse}')
print(f'RMSE={rmse}')
print(f'R2={r2}')  # 설명력
```

```python
y_train_pred = model.predict(X_train)  # train값으로 예측시키기
train_rmse = root_mean_squared_error(y_train, y_train_pred)

print(f'훈련 세트 RMSE: {train_rmse}')
print(f'테스트 세트 RMSE: {rmse}')
```

- 훈련세트와 테스트세트의 RMSE가 거의 차이 없이 나타났습니다.
- 모델이 훈련 데이터의 과도하게 맞춰지지 않고, (과적합 없이)
- 새로운 데이터에도 비슷한 오차 수준을 유지하며
- 잘 인반화 된 모델이라고 판단할 수 있습니다.

- 위도, 경도까지 포함해서 예측하는게 맞을까?
    - 제거를 하고 다시 예측모델을 돌려보자

```python
# 학습/테스트 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=95)
```

```python
# 모델 학습
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# RMSE, R2 평가
y_pred = model.predict(X_test)  # test값으로 예측시키기

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'MSE={mse}')
print(f'RMSE={rmse}')
print(f'R2={r2}')
```
